```python
import pinecone
from sentence_transformers import SentenceTransformer
from pinecone import ServerlessSpec

# Load the sentence-transformers model
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Initialize Pinecone
pc = pinecone.Pinecone(api_key="use your own api key")

index_name = "ui-components"

# Use us-east-1 which is supported in Free Plan on AWS
if index_name not in [index.name for index in pc.list_indexes()]:
    pc.create_index(
        name=index_name,
        dimension=384,
        metric="cosine",
        spec=ServerlessSpec(
            cloud='aws',
            region='us-east-1'
        )
    )

# Connect to the Pinecone index
index = pc.Index(index_name)

# Generate embeddings and upload to Pinecone
vectors = []
for i, component in enumerate(components):
    text_to_embed = f"{component['name']} {component['description']} {', '.join(component['tags'])} {component['code']}"
    component_embedding = model.encode(text_to_embed).tolist()

    metadata = {
        "name": component["name"],
        "description": component["description"],
        "tags": component["tags"],
        "source": component["source"]
    }

    vectors.append({
        "id": str(i),
        "values": component_embedding,
        "metadata": metadata
    })

index.upsert(vectors=vectors)
```

```python
query_text = "A table with three rows"
query_embedding = model.encode(query_text)

query_embedding = query_embedding.tolist()

query_results = index.query(
    vector=query_embedding,
    top_k=3,
    include_metadata=True
)

# Print similar components
for match in query_results["matches"]:
    print(f"Similar Component: {match['metadata']['name']}")
```
